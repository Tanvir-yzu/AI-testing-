```python
import gradio as gr
```

```python
def hello(name: str) -> str:
    if not isinstance(name, str):
        raise TypeError("Name must be a string")
    if not name.strip():
        raise ValueError("Name cannot be empty")
        
    return f"Hello {name.strip()}!"
```

```python
# The correct way to create a Gradio interface
inter = gr.Interface(
    fn=hello,
    inputs=gr.Textbox(
        label="Input Text",
        placeholder="Enter your text here...",
        lines=2
    ),
    outputs=gr.Textbox(
        label="Output Text",
        lines=2
    )
)
```

```python
inter.launch()
```

```python
import gradio as gr
import numpy as np

def sepia(input_img):
    # Check if input image is not None
    if input_img is None:
        return None
        
    # Convert input to float32 for better precision
    input_img = input_img.astype(np.float32)
    
    sepia_filter = np.array([[.393, .769, .189],
                            [.349, .686, .168],  # Fixed spacing typo
                            [.272, .534, .131]])
    
    # Ensure input image has correct dimensions
    if len(input_img.shape) < 3:
        return input_img
        
    sepia_img = input_img.dot(sepia_filter.T)
    
    # Clip values to valid range [0, 1]
    sepia_img = np.clip(sepia_img, 0, 1)
    
    return sepia_img

# Create interface with proper output type
iface = gr.Interface(
    sepia,
    inputs=gr.Image(width=200, height=200),
    outputs=gr.Image(),
    title="Sepia Filter"
)

iface.launch()
```
